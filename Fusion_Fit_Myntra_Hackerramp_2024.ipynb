```python
def combine_recommendations(cf_result,cbf_results,trend_results):
  combined_df=pd.DataFrame(columns=['product_id','cf_score','cbf_score','trend_score'])

  for product_id in cf_result:
    if product_id in combined_df['product_id'].values:
      combined_df.loc[combined_df['product_id']==product_id,'cf_score']+=1
    else:
      combined_df=combined_df.append({'product_id':product_id,'cf_score':1,'cbf_score':0,'trend_score':0},ignore_index=True)

  for product_id in cbf_result:
    if product_id in combined_df['product_id'].values:
      combined_df.loc[combined_df['product_id']==product_id,'cbf_score']+=1
    else:
      combined_df=combined_df.append({'product_id':product_id,'cf_score':0,'cbf_score':1,'trend_score':0},ignore_index=True)

  for product_id in trend_results:
    if product_id in combined_df['product_id'].values:
      combined_df.loc[combined_df['product_id']==product_id,'trend_score']+=1
    else:
      combined_df=combined_df.append({'product_id':product_id,'cf_score':0,'cbf_score':0,'trend_score':1},ignore_index=True)

  combined_df['total_score']=combined_df['trend_score']+combined_df['cf_score']+combined_df['cbf_score']

  combined_df=combined_df.sort_values(by='total_score',ascending=False)

  combine_recommendations=combined_df['product_id'].tolist()

  return combine_recommendations
```

```python
import pandas as pd
from IPython.display import Image, display

def recommend_accessories_with_images(csv_file_path):
    # Load the CSV file
    accessory_df = pd.read_csv(csv_file_path)

    if 'ratingCount' not in accessory_df.columns:
        raise ValueError("Column 'ratingCount' not found in the CSV file.")

    accessory_df['ratingCount'] = pd.to_numeric(accessory_df['ratingCount'], errors='coerce')  # Convert to numeric

    top_accessories = accessory_df.sort_values(by='ratingCount', ascending=False).head(3)

    top_accessories_info = []

    for index, row in top_accessories.iterrows():
        product_id = row['p_id']
        image_url = row['img']
        top_accessories_info.append((image_url, product_id, row["ratingCount"]))

    return top_accessories_info
```

```python
def initialize_csv(file_path):
    try:
        df = pd.read_csv(file_path)
    except FileNotFoundError:
        df = pd.DataFrame(columns=['customer_id', 'liked_products', 'disliked_products'])
        df.to_csv(file_path, index=False)
```

```python
def load_user_interactions(file_path):
    try:
        df = pd.read_csv(file_path)
        return df
    except FileNotFoundError:
        return pd.DataFrame(columns=['customer_id', 'liked_products', 'disliked_products'])
```

```python
def save_user_interaction(customer_id, liked_products, disliked_products, file_path):
    user_interaction = load_user_interactions(file_path)

    new_interaction = pd.DataFrame({
        'customer_id': [customer_id],
        'liked_products': [','.join(liked_products)],
        'disliked_products': [','.join(disliked_products)]
    })
    user_interaction = pd.concat([user_interaction, new_interaction], ignore_index=True)
    user_interaction.to_csv(file_path, index=False)
    return user_interaction
```

```python
def update_user_profile(customer_id, feedback, file_path, product_map):
    global user_profiles
    if customer_id not in user_profiles:
        user_profiles[customer_id] = {
            'liked_products': [],
            'disliked_products': [],
            'rating_changes': {}
          }

    for liked_product in feedback['liked_products']:
        product_ids = product_map.get(liked_product, [])
        user_profiles[customer_id]['liked_products'].extend(product_ids)
        for product_id in product_ids:
            save_user_interaction(customer_id, product_id, liked_product, None, file_path)

    for disliked_product in feedback['disliked_products']:
        product_ids = product_map.get(disliked_product, [])
        user_profiles[customer_id]['disliked_products'].extend(product_ids)
        for product_id in product_ids:
            save_user_interaction(customer_id, product_id, None, disliked_product, file_path)

    user_profiles[customer_id]['rating_changes'].update(feedback['rating_changes'])

    return "User profile updated successfully"
```

```python
def find_clothes(product_id, product_description, default_image):
    accessory_image = product_description.loc[product_description['p_id'] == product_id, 'img'].values
    if len(accessory_image) > 0:
        return accessory_image[0]
    return default_image
```

```python
def recommended_clothes(customer_id, product_description, user_interaction, default_image):
    if 'customer_id' not in user_interaction.columns or 'p_id' not in user_interaction.columns:
        print("User interaction DataFrame is missing required columns.")
        return []

    recommended_products = user_interaction[user_interaction['customer_id'] == customer_id]['p_id'].tolist()
    recommendations = []
    for product_id in recommended_products:
        image_url = find_clothes(product_id, product_description, default_image)
        recommendations.append((product_id, image_url))
    return recommendations
```

```python
import pandas as pd
from IPython.display import Image, display
def recommend_products(user_id, product_description, liked_products, num_recommendations=3):
    liked_products_lower = [prod.lower() for prod in liked_products]
    relevant_products = product_description[product_description['products'].str.lower().str.contains('|'.join(liked_products_lower))]
    if not relevant_products.empty:
        recommended_products = relevant_products.sample(min(num_recommendations, len(relevant_products)))
    else:
        recommended_products = product_description.sample(num_recommendations)
    product_ids = recommended_products['p_id'].tolist()
    product_names = recommended_products['products'].tolist()
    image_urls = recommended_products['img'].tolist()

    return product_ids, product_names, image_urls

def display_recommendations(product_names, image_urls):
    for name, url in zip(product_names, image_urls):
        print(f"Product: {name}")
        display(Image(url=url))

def main():
    user_interaction_file = '/content/user_interactions.csv'
    initialize_csv(user_interaction_file)
    product_description = pd.read_csv('/content/product_dataset.csv')
    balanced_product_dataset = pd.read_csv('/content/balanced_product_dataset.csv')
    default_image = '/content/default.jpg'

    while True:
        print("\nWelcome to the Recommendation System!")
        print("1. Update User Profile")
        print("2. Recommend Clothes")
        print("3. Top Rated Accessories")
        print("4. Retrieve and Display Product Info")
        print("5. Exit")

        choice = input("Enter your choice (1/2/3/4/5): ")

        if choice == '1':
            print("\nUpdate User Profile:")
            customer_id = int(input("Enter Customer ID: "))

            liked_products_input = input("Enter liked product names (comma-separated): ").split(',')
            liked_products = [prod.strip() for prod in liked_products_input]

            disliked_products_input = input("Enter disliked product names (comma-separated): ").split(',')
            disliked_products = [prod.strip() for prod in disliked_products_input]

            user_interaction = save_user_interaction(customer_id, liked_products, disliked_products, user_interaction_file)
            print("User profile updated successfully.")
            print("Current User Interactions:")
            print(user_interaction)

        elif choice == '2':
            print("\nRecommend Clothes:")
            user_interaction = pd.read_csv(user_interaction_file)

            if user_interaction.empty:
                print("No user interactions found. Please update a profile first.")
                continue

            customer_id = int(input("Enter Customer ID to get recommendations: "))
            liked_products = user_interaction[user_interaction['customer_id'] == customer_id]['liked_products'].tolist()

            if not liked_products:
                print("No liked products found for this user. Please update your profile.")
                continue
            product_ids, product_names, image_urls = recommend_products(customer_id, product_description, liked_products)

            if product_names:
                display_recommendations(product_names, image_urls)
                selection = int(input("Enter the number of the product you want to choose: "))
                if 1 <= selection <= len(product_names):
                    selected_product_name = product_names[selection - 1]
                    print(f"Selected Product Name: {selected_product_name}")

                else:
                    print("Invalid selection. Please try again.")
            else:
                print("No recommendations found for this user.")

        elif choice == '3':
            print("\nTop Rated Accessories:")
            try:
                top_accessories_info = recommend_accessories_with_images('/content/accessory.csv')
                for img, p_id, ratingCount in top_accessories_info:
                    display(Image(url=img, width=200))
                    print(f"Product ID: {p_id}")
                    print(f"Rate Count: {ratingCount}")

            except ValueError as e:
                print(str(e))

        elif choice == '4':
            print("\nRetrieve and Display Product Info:")
            customer_id = int(input("Enter Customer ID to retrieve product info: "))
            retrieve_and_display_product_info(customer_id, product_description, balanced_product_dataset, default_image)

        elif choice == '5':
            print("Exiting...")
            break

        else:
            print("Invalid choice. Please enter 1, 2, 3, 4, or 5.")

if __name__ == "__main__":
    main()
```
